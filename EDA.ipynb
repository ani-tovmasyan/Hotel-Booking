# Distribution Analysis:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter outliers for ADR (focusing on a more common range)
filtered_adr = hotel_bookings['adr'][(hotel_bookings['adr'] > 0) & (hotel_bookings['adr'] < 1000)]

# Filter stay_duration to a more common range to avoid extreme values
filtered_stay_duration = hotel_bookings['stay_duration'][(hotel_bookings['stay_duration'] > 0) & (hotel_bookings['stay_duration'] <= 30)]

# Setup for adjusted visualizations
plt.figure(figsize=(18, 6))

# Lead Time
plt.subplot(1, 3, 1)
sns.histplot(hotel_bookings['lead_time'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (days)')
plt.ylabel('Count')

# ADR (Average Daily Rate)
plt.subplot(1, 3, 2)
sns.histplot(filtered_adr, bins=40, kde=True, color='lightgreen')
plt.title('Distribution of ADR (Filtered)')
plt.xlabel('ADR (€)')
plt.ylabel('Count')

# Stay Duration
plt.subplot(1, 3, 3)
sns.histplot(filtered_stay_duration, bins=30, kde=True, color='salmon')
plt.title('Distribution of Stay Duration (Filtered)')
plt.xlabel('Stay Duration (nights)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

* Lead Time: The distribution is likely skewed to the right, meaning most bookings are made relatively close to the arrival date, but there are outliers with very long lead times.
* Average Daily Rate (ADR): This distribution could also be right-skewed, indicating that most bookings fall into a lower price range, with fewer bookings at very high prices.
* Stay Duration: Expected to be right-skewed as well, with most stays being short and fewer long stays.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Creating subplots for the histograms of numerical variables
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Distribution of lead_time
sns.histplot(hotel_bookings['lead_time'], bins=50, kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Lead Time')

# Distribution of adr (Average Daily Rate)
sns.histplot(hotel_bookings['adr'], bins=50, kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Distribution of Average Daily Rate (ADR)')

# Distribution of stays_in_weekend_nights
sns.histplot(hotel_bookings['stays_in_weekend_nights'], bins=30, kde=False, ax=axs[1, 0])
axs[1, 0].set_title('Distribution of Stays in Weekend Nights')

# Distribution of stays_in_week_nights
sns.histplot(hotel_bookings['stays_in_week_nights'], bins=30, kde=False, ax=axs[1, 1])
axs[1, 1].set_title('Distribution of Stays in Week Nights')

plt.tight_layout()
plt.show()
```
